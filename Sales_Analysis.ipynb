## Objective
Analyze sales transaction data and derive business insights using Pandas.

```python
import pandas as pd

# Sales Data
sales_data = {
    'order_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110, 111, 112],
    'customer': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob', 'Alice', 'David', 'Charlie', 'Bob', 'Alice', 'David', 'Charlie'],
    'product': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Mouse', 'Monitor', 'Laptop', 'Mouse', 'Keyboard', 'Laptop', 'Monitor', 'Keyboard'],
    'category': ['Electronics']*12,
    'amount': [50000, 500, 50000, 1500, 500, 12000, 50000, 500, 1500, 50000, 12000, 1500],
    'city': ['Mumbai', 'Delhi', 'Mumbai', 'Bangalore', 'Delhi', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai']
}

df = pd.DataFrame(sales_data)
df
```

## Task 1: Value Counts and Sorting

```python
# Part A
df['customer'].value_counts()

# Part B
df['product'].value_counts().head(3)
```

### Explanation:
- value_counts() counts frequency of unique values.
- ascending=True sorts low to high, False sorts high to low.

## Task 2: Handling Duplicates

```python
# Part A
df_no_duplicates = df.drop_duplicates(subset=['customer', 'product'], keep='first')
df_no_duplicates

# Part B
df['city'].unique()
```

### Explanation:
- subset specifies columns to check duplicates.
- keep='first' keeps first occurrence.
- inplace=True modifies original dataframe.

## Task 3: GroupBy Operations

```python
# Part A
df.groupby('customer')['amount'].sum().sort_values(ascending=False)

# Part B
df.groupby('city')['amount'].sum()

# Part C
df.groupby('customer')['product'].nunique()
```

## Insights
- Alice is the highest spender.
- Laptop is the most sold product.
- Mumbai has the highest sales.
